```python
import numpy as np

def cap_outliers_iqr(series):
    q1 = np.percentile(series, 25)
    q3 = np.percentile(series, 75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series.clip(lower=lower, upper=upper)

important_cols = ['Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber',
                  'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content', 'Serving_Size']

for col in important_cols:
   df.loc[:, col] = cap_outliers_iqr(df[col])

# Cap outliers in important columns using IQR method
```

```python
df[important_cols].describe()
# Descriptive statistics after capping outliers in numerical columns
```

```python
from sklearn.impute import SimpleImputer

# Columns you want to impute
important_cols = ['Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber',
                  'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content', 'Serving_Size']

# Create median imputer
imputer = SimpleImputer(strategy='median')

# Apply imputer and assign back to the DataFrame
df.loc[:, important_cols] = imputer.fit_transform(df[important_cols])

```
